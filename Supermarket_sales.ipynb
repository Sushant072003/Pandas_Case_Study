**9. Display Daily Sales by Day of the Week**

```python
df.columns
```

```python
dw_mapping={
    0:"Mon",
    1:"Tue",
    2:"Wed",
    3:"Thur",
    4:"Fri",
    5:"Sat",
    6:"Sun"
}
```

```python
df['day_of_week'] = df['Date'].dt.day_of_week.map(dw_mapping)
```

```python
df['day_of_week'].head(2)
```

```python
df['day_of_week'].value_counts().plot(kind='bar')
```

**10. What will be the Highest Months for Sales**

```python
dm_mapping={
    1:"Jan",
    2:"Feb",
    3:"Mar"
}
```

```python
df['Months'] = df['Date'].dt.month.map(dm_mapping)
```

```python
df['Months'].value_counts().plot(kind='bar')
```
